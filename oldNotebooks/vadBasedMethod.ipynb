```python
import scipy.io.wavfile
import numpy as np
from IPython.display import display, Audio
import librosa
import matplotlib.pyplot as plt

rate = 16000
SAMPLE_START = 0
SPEECH_START_BAND = 300
SPEECH_END_BAND = 3000
SAMPLE_WINDOW = 0.02*rate
SPEECH_WINDOW = 0.5
SAMPLE_OVERLAP = 0.01
THRESHOLD = 0.6
```

```python
from scipy.signal import butter, lfilter

wav, sr = librosa.load('new.wav')

# Define the bandpass filter parameters
lowcut = 300  # Lower cutoff frequency in Hz
highcut = 5400  # Upper cutoff frequency in Hz
order = 6  # Filter order

# Create a Butterworth bandpass filter
nyquist_freq = 0.5 * sr
low = lowcut / nyquist_freq
high = highcut / nyquist_freq
b, a = butter(order, [low, high], btype='band')

# Apply the bandpass filter to the audio signal
filtered_signal = lfilter(b, a, wav)

display(Audio(filtered_signal, rate=sr))
```

```python

S_wav, phase = librosa.magphase(librosa.stft(wav))

S_new = S_wav

for i in range(0, len(S_wav)):
    if np.mean(S_wav[i])>0.6:
        S_new[i] = np.mean(S_new[i])*np.ones(S_new[i].shape)

y_new = librosa.istft(S_new*phase)

display(Audio(y_new, rate=sr))

plt.figure(figsize=(10,7))
librosa.display.specshow(librosa.power_to_db(S_new, ref=np.max), y_axis='log', sr=sr)
plt.colorbar()
```
